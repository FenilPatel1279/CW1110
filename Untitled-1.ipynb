```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
import numpy as np

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Initialize KNN
knn = KNeighborsClassifier(n_neighbors=3)

# Create KFold with shuffle enabled
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform K-Fold cross-validation
scores = cross_val_score(knn, X, y, cv=kf)

print("K-Fold Cross-validation scores:", scores)
print("Average Accuracy:", np.mean(scores))
```

```python
import matplotlib.pyplot as plt

folds = range(1, 6)

plt.figure(figsize=(8, 5))
plt.bar(folds, scores)
plt.title("Cross-Validation Accuracy per Fold")
plt.xlabel("Fold Number")
plt.ylabel("Accuracy")
plt.ylim(0.7, 1.0)
plt.grid(True)
plt.show()
```

```python
# Import libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np

# Load dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = X['petal length (cm)']
X = X.drop('petal length (cm)', axis=1)

# Initialize model
model = DecisionTreeRegressor(random_state=42)

# Define custom scorer for negative MSE
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)

# Set up explicit K-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
cv_scores = cross_val_score(model, X, y, cv=kf, scoring=mse_scorer)

# Convert to positive MSE for readability
mse_scores = -cv_scores
print("K-Fold Cross-Validation MSE scores:", mse_scores)
print("Average MSE:", np.mean(mse_scores))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample CV MSE scores (use your actual output here)
# mse_scores = -cross_val_score(model, X, y, cv=5, scoring=mse_scorer)

folds = np.arange(1, len(mse_scores) + 1)

# Bar Plot of MSE per Fold
plt.figure(figsize=(8, 5))
plt.bar(folds, mse_scores)
plt.xlabel('Fold Number')
plt.ylabel('Mean Squared Error')
plt.title('Cross-Validation MSE per Fold (Decision Tree)')
plt.grid(True)
plt.show()

# Line Plot of MSE per Fold
plt.figure(figsize=(8, 5))
plt.plot(folds, mse_scores, marker='o', linestyle='--')
plt.xlabel('Fold Number')
plt.ylabel('Mean Squared Error')
plt.title('Cross-Validation MSE Trend')
plt.grid(True)
plt.show()
```

```python
avg_mse = mse_scores.mean()

# Re-plot with average line
plt.figure(figsize=(8, 5))
plt.bar(folds, mse_scores, color='skyblue')
plt.axhline(avg_mse, color='red', linestyle='--', label=f'Avg MSE = {avg_mse:.2f}')
plt.xlabel('Fold Number')
plt.ylabel('Mean Squared Error')
plt.title('Cross-Validation MSE with Average Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Initialize the model
knn = KNeighborsClassifier(n_neighbors=3)

# Perform 5-fold cross-validation
cv_scores = cross_val_score(knn, X, y, cv=5)

# Print results
print("Cross-validation accuracy scores:", cv_scores)
print("Average accuracy:", np.mean(cv_scores))
```
